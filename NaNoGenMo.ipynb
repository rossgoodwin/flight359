```
JSON_DIR = '/content/drive/My Drive/flight359'


json_fps = map( lambda x: os.path.join(JSON_DIR, x), filter(lambda y: y.endswith('.json'), os.listdir(JSON_DIR)) )

def chapter_txt(chapter_name):
	return "\\chapter{%s}" % chapter_name

def process_json(fp):
  with open(fp) as infile:
    obj = json.load(infile)

  poem_title = obj['title']
  poem_txt = obj['poem']
  raw_lines = poem_txt.replace("&", "\\&").split('\n')
  first_line = raw_lines[0]
  second_forward = raw_lines[1:]

  return "%s\n%s\n%s" % (
    chapter_txt(poem_title),
    first_line,
    '\n'.join( map(lambda x: "\\\\%s" % x.strip().strip('[').strip(']'), second_forward ) )
  )


book_txt = "\n\n\n\n".join( map( process_json, json_fps ) )

with open(os.path.join(JSON_DIR, 'raw_tex_v2.txt'), 'w') as outfile:
  outfile.write(book_txt)
```
